```python
name = str(input("Enter student's name: "))

def analyze_student_score(passing_score=75):
  user_list = []
  try:
    count = int(input(f"Enter number of quizzes taken for {name}? "))
    if count <= 0:
      print("Please enter a valid number of quizzes (greater than 0).")
      return # Exit if count is invalid

    for i in range(count):
      while True:
        try:
          val = float(input(f"Enter score {i + 1}: "))
          user_list.append(val)
          break
        except ValueError:
          print("Invalid input. Please enter a valid number.")

  except ValueError: # Catch ValueError from int(input) for 'count'
    print("Invalid input for number of quizzes. Please enter a valid integer.")
    return

  # Check if user_list is empty after attempts to get scores
  if not user_list:
     print("No valid scores were entered or processed.")
     return

  total_score = sum(user_list)
  average_score = total_score / len(user_list)

  if average_score >= passing_score:
    print(f"Result: Passed ")
  else:
    print(f"Result: Failed ")

analyze_student_score()
```
